```python
for i in ['Duration to accept offer','Notice period','Pecent hike expected in CTC','Percent hike offered in CTC',
          'Percent difference CTC','Rex in Yrs','Age']: 
    plt.figure()
    plt.boxplot(data1[i])
    plt.title(i)
```

```python
def iqr_capping(data1,variables):
    q1=data1[variables].quantile(0.25)
    q3=data1[variables].quantile(0.75)
    iqr=q3-q1
    upper_whisker=q3+(1.5*iqr)
    lower_whisker=q1-(1.5*iqr)
    return lower_whisker,upper_whisker
```

```python
lower_lim,upper_lim=iqr_capping(data1,'Duration to accept offer')
print('Duration to accept offer= ',lower_lim)
print('Duration to accept offer= ',upper_lim)
```

```python
data1['Duration to accept offer']=np.where(data1['Duration to accept offer']>upper_lim,upper_lim,
                    np.where(data1['Duration to accept offer']<lower_lim,lower_lim,data1['Duration to accept offer']))
```

```python
lower_lim,upper_lim=iqr_capping(data1,'Notice period')
print('Notice period= ',lower_lim)
print('Notice period= ',upper_lim)
data1['Notice period']=np.where(data1['Notice period']>upper_lim,upper_lim,
                    np.where(data1['Notice period']<lower_lim,lower_lim,data1['Notice period']))
```

```python
lower_lim,upper_lim=iqr_capping(data1,'Pecent hike expected in CTC')
print('Pecent hike expected in CTC= ',lower_lim)
print('Pecent hike expected in CTC= ',upper_lim)
data1['Pecent hike expected in CTC']=np.where(data1['Pecent hike expected in CTC']>upper_lim,upper_lim,
                    np.where(data1['Pecent hike expected in CTC']<lower_lim,lower_lim,data1['Pecent hike expected in CTC']))
```

```python
lower_lim,upper_lim=iqr_capping(data1,'Percent hike offered in CTC')
print('Percent hike offered in CTC= ',lower_lim)
print('Percent hike offered in CTC= ',upper_lim)
data1['Percent hike offered in CTC']=np.where(data1['Percent hike offered in CTC']>upper_lim,upper_lim,
                    np.where(data1['Percent hike offered in CTC']<lower_lim,lower_lim,data1['Percent hike offered in CTC']))
```

```python
lower_lim,upper_lim=iqr_capping(data1,'Percent difference CTC')
print('Percent difference CTC= ',lower_lim)
print('Percent difference CTC= ',upper_lim)
data1['Percent difference CTC']=np.where(data1['Percent difference CTC']>upper_lim,upper_lim,
                    np.where(data1['Percent difference CTC']<lower_lim,lower_lim,data1['Percent difference CTC']))
```

```python
lower_lim,upper_lim=iqr_capping(data1,'Rex in Yrs')
print('Rex in Yrs= ',lower_lim)
print('Rex in Yrs= ',upper_lim)
data1['Rex in Yrs']=np.where(data1['Rex in Yrs']>upper_lim,upper_lim,
                    np.where(data1['Rex in Yrs']<lower_lim,lower_lim,data1['Rex in Yrs']))
```

```python
lower_lim,upper_lim=iqr_capping(data1,'Age')
print('Age= ',lower_lim)
print('Age= ',upper_lim)
data1['Age']=np.where(data1['Age']>upper_lim,upper_lim,
                    np.where(data1['Age']<lower_lim,lower_lim,data1['Age']))
```

# over sampling

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
X1,Y1=oversample.fit_resample(X,Y)
```

```python
#checking the oversampling output
Y2=pd.DataFrame(Y1)
Y2.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X1,Y1,random_state=42,test_size=.2)
```

# Scaling

```python
# minmax scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler(feature_range=(0,1))
X_train=min_max_scaler.fit_transform(X_train)
X_test=min_max_scaler.fit_transform(X_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression(random_state=42)
logit_model.fit(X_train,Y_train)
Y_pred=logit_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print('Accuracy is',accuracy_score(Y_test,Y_pred))
```

```python
Accuracy_log=accuracy_score(Y_test,Y_pred)
print(Accuracy_log)
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp=ConfusionMatrixDisplay(confusion_matrix(Y_test,Y_pred))
disp.plot()
plt.show
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k=[]
neighbors=np.arange(3,15)
for k in neighbors:
    classifier= KNeighborsClassifier(n_neighbors=k,metric='minkowski',p=2)
    classifier.fit(X_train,Y_train)
    Y_pred=classifier.predict(X_test)
    acc=accuracy_score(Y_test,Y_pred)
    metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```

```python
classifier=KNeighborsClassifier(n_neighbors=2,metric='minkowski',p=2)
classifier.fit(X_train,Y_train)
Y_pred=classifier.predict(X_test)
```

```python
print('Accuracy is',accuracy_score(Y_test,Y_pred))
confusion_matrix(Y_test,Y_pred)
```

```python
Accuracy_knn=accuracy_score(Y_test,Y_pred)
print(Accuracy_knn)
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix(Y_test,Y_pred))
disp.plot()
plt.show
```
